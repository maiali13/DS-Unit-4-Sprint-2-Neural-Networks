### Input Layer:
the "visible layer". Includes all of the possible input nodes. Input nodes have a 1 to 1 relationship with the number of features in a dataset.

### Hidden Layer:
The nidden nodes between the input and output nodes, just a step in the process of getting final output. 

### Output Layer:
Output nodes. Essentially a value or class (target). The total number changes depending on the type of problem the NN is being constructed to solve. Linear regression problems onbly need 1 node while classification problems will have the same number of output modes as possible classes. 

### Neuron:
ML terms node. The neurons of the input layer are the initial inputs, they are not calculated.

### Weight:
Values of hidden nodes a value that designate the importance of one neuron in the calculation of a neuron in the next layer. Similar to coefficients. Hodden nodes "weight" the input by x amount and then pass that output to next layer of nodes.

### Activation Function:
"transfer functions" used in NNs to standardize neuron values between 0 and 1. (for example Sigmoid, Tanh, or Relu)
introduces non-linearlity to the NN 

### Node Map:
A visualization a NN including the neurons and the connections between them. 

### Perceptron:
The simplest neural network, with a vector input and single output.

### Explain the flow of information through a neural network from inputs to outputs. Be sure to include: inputs, weights, bias, and activation functions. How does it all flow from beginning to end?

#### 
NNs process complex input data like audio, text, or images. First the image is given to the input node, which reduce the dimensionality of the data by, for example, displaying each pixel via a RGB value or a brightness value between 0 and 1. This is then passed to the hidden nodes, which apply a weight to the data they recieve, and then applyes a bias (+/-) to the sum of that weight. Across the hidden node neurons, these values then get normalized by the activation function (to be between 0 and 1) and then get passed to the next layer until the data reaches the output layer. In this manner, the information which is output by the NN has lower dimensionality than the input, either as a value or classification, and is more interpertable.

```python
##### Update this Class #####

class Perceptron(object):
    """
    Pereceptrion class to fit and predict
    """
    def __init__(self, niter = 10):
        """default bias = 10 and n_iter """
        self.niter = niter
    
    def __sigmoid(self, x):
        """convert to float between 0-1"""
        return 1 / (1 + np.exp(-x))
    
    def __sigmoid_derivative(self, x):
        """returns the derivite of the sigmoid value """
        sx = sigmoid(x)
        return sx * (1-sx)

    def fit(self, X, y):
        """Fit training data
        X : Training vectors, X.shape : [features]
        y : Target values, y.shape : [target]
        """
        input = X
                
        # Randomly Initialize Weights
        weights =  np.random.rand(X.shape[1], 1)

        for i in range(self.niter):
            # Weighted sum of inputs / weights
            weighted_sum = np.dot(input, weights) + 5
            # Activate!
            self.activated_output = self.__sigmoid(weighted_sum)
            # Cac error
            error = y - self.activated_output
            # Update the Weights
            adjustments = error * sigmoid_derivative(self.activated_output)
            weights += np.dot(X.T, adjustments)
        return self

    def predict(self, X):
        """Return class label after unit step"""
        return self.activated_output
```

```python
# use the class!
scaler = MinMaxScaler()
X = diabetes[feats]
y = [diabetes['Outcome']]
y_arr = scaler.fit_transform(y).T
X = np.array(X.values.tolist())

ptron = Perceptron()
ptron.fit(X,y_arr)
```
